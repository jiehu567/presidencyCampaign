```python
from __future__ import print_function
import os
import six
import rake
import operator
import io

# import sys
# reload(sys)
# sys.setdefaultencoding('utf-8')
```

```python
# RAKE method
def keywordExtract(filename):
    
    # EXAMPLE ONE - SIMPLE
    stoppath = "SmartStoplist.txt"
    
    
    # print("---------------------------------------------")
    # EXAMPLE TWO - BEHIND THE SCENES (from https://github.com/aneesha/RAKE/rake.py)
    
    # 1. initialize RAKE by providing a path to a stopwords file
    # rake_object = rake.Rake(stoppath)
    
    sample_file = io.open(filename, 'r',encoding="iso-8859-1")
    text = sample_file.read()
    
    # 1. Split text into sentences
    sentenceList = rake.split_sentences(text)
    
    # generate candidate keywords
    stopwordpattern = rake.build_stop_word_regex(stoppath)
    phraseList = rake.generate_candidate_keywords(sentenceList, stopwordpattern)
    # print("Phrases:", phraseList)
    # print("---------------------------------------------")
    
    # calculate individual word scores
    wordscores = rake.calculate_word_scores(phraseList)
    
    # generate candidate keyword scores
    keywordcandidates = rake.generate_candidate_keyword_scores(phraseList, wordscores)
    
    # sort candidates by score to determine top-scoring keywords
    sortedKeywords = sorted(six.iteritems(keywordcandidates), key=operator.itemgetter(1), reverse=True)
    # totalKeywords = len(sortedKeywords)
    return sortedKeywords
    # print("---------------------------------------------")
    # for example, you could just take the top third as the final keywords
    # for keyword in sortedKeywords[0:int(totalKeywords / 3)]:
    #    print "Keyword: %s\tscore: %s" % (keyword[0], keyword[1])
    
    # print(rake_object.run(text))
```

```python
def outputKeyword():
    dir = os.getcwd() + "/" + "HillarySpeeches"
    
    words_lsts = []
    
    for filename in os.listdir(dir):
        if filename.startswith("1") == False: continue
    
        dir_file = dir + '/' + filename
        sortedkeywords = keywordExtract(dir_file)
        
        word_lst = []
        for keyword in sortedkeywords[0:20]:   # int(totalKeywords / 3)
            
            keyword_0 = keyword[0].split()
            
            for kw in keyword_0:
                word_lst.append(kw)

        words_lsts.append(word_lst)        
        
    return words_lsts
```

#### Pre-processing word list
- remove: keywords with numbers, punctuations
- transform into normal string
- remove wrong unicode words

```python
import string
def hasNumbersOrPunc(inputString):
    return any(char.isdigit() or char in string.punctuation for char in inputString)
```

```python
wl = outputKeyword()
```

```python
wl[1][:10]
```

```python
# wl = [[word.encode('utf-8').strip() for word in w] for w in wl]
wl = [[word for word in row if hasNumbersOrPunc(word) is False] for row in wl]

for row in wl:
    for word in row:
        if "\\" in r"%r" % word:
            row.remove(word)
```

```python
wl
```

```python
## transform keywords into their stem words
from nltk.stem.snowball import EnglishStemmer

# only get stem words of each article's word list

es = EnglishStemmer()

stem_wl = [[es.stem(word.lower()) for word in w] for w in wl]

# save all stem words in a list for vectorizing later
big_wordlst = []

for each_lst in wl:
    for word in each_lst:
        append_word =  es.stem(word.lower())
        if append_word not in big_wordlst:
            big_wordlst.append(append_word)
            
# size of big_word_lst: 1227
len(big_wordlst)

# see how sorted big_word_lst looks like:
big_wordlst.sort()
```

```python
big_wordlst
```

```python
# Create data frame
# column names (features) are stemmed keywords in big_wdlst
# rows are documents index

import pandas as pd
import numpy as np

columns = big_wordlst
rows = range(len(wl))
data = np.zeros((len(wl),len(columns)))

df = pd.DataFrame(data = data,index=rows, columns=columns)

df.head(5)[df.columns[0:10]]

# get frequency of stemmed keywords in each article
for article in range(len(stem_wl)):
    for word in stem_wl[article]:
        if word in big_wordlst:
            data[article][big_wordlst.index(word)] += 1

# Now we have matrix/data frame of each row as frequency vector of an article
df = pd.DataFrame(data = data,index=rows, columns=columns)
# see how many non-zeros in this data frame
# count = pd.DataFrame(data = len(data[1]) - (df == 0).astype(int).sum(axis=1),index=rows)
# ratio = pd.DataFrame(data = count/len(data[1]),index=rows)

# pd.concat([count, ratio],axis=1)
```

```python
# use k-means to cluster

from sklearn.cluster import KMeans

num_clusters = 3

km = KMeans(n_clusters=num_clusters)

km.fit(data)

clusters = km.labels_.tolist()

speeches = {'doc_index':rows,'keywords':wl, 'cluster': clusters}

frame = pd.DataFrame(speeches, index = [clusters] , columns = ['doc_index', 'cluster', 'keywords'])
frame

frame['cluster'].value_counts()
```

```python
# save doc's indexes of each cluster
docCluster_index = [[],[],[]]
for doc_index in range(len(clusters)):
    docCluster_index[clusters[doc_index]].append(doc_index)
```
